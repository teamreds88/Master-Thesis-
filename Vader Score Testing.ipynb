```python
############# TESTING EMOJI SCORING WITH VADER ###########################
!pip install vaderSentiment
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import pandas as pd

# Initialize VADER analyzer
analyzer = SentimentIntensityAnalyzer()

# Sample emoji-containing comments
comments = [
    "I absolutely loved this movie! 😍🎉",
    "This was terrible... I want my time back 😡💩",
    "Just watched it. 🤔 Not sure how I feel.",
    "It was okay. Nothing special. 🙂",
    "Great acting but poor storyline. 🤷‍♂️",
    "💯🔥 Must-watch for everyone!",
    "Ugh. Waste of time 😤",
    "Wow! I'm speechless 😭❤️",
    "No words. Just... wow. 😳",
    "meh. 😐"
]

# Analyze sentiment
results = []
for comment in comments:
    score = analyzer.polarity_scores(comment)
    results.append({
        'comment': comment,
        'compound': score['compound'],
        'pos': score['pos'],
        'neu': score['neu'],
        'neg': score['neg']
    })

# Create and display DataFrame
emoji_test_scores = pd.DataFrame(results)
print(emoji_test_scores)
```

```python
########### testing how sentiment scores change as more and more emojis are added in
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()

test_sentences = {
    "positive": [
        "I love this product",
        "I love this product ❤️",
        "I love this product ❤️❤️❤️"
    ],
    "negative": [
        "This was the worst experience",
        "This was the worst experience 😡",
        "This was the worst experience 😡😡😡"
    ],
    "neutral": [
        "I went to the store",
        "I went to the store 🛒",
        "I went to the store 🛒🛒🛒"
    ]
}

# Score each version
for label, versions in test_sentences.items():
    print(f"\n--- {label.upper()} SENTENCE TEST ---")
    for sentence in versions:
        score = analyzer.polarity_scores(sentence)
        print(f"{sentence} --> Compound: {score['compound']}, Pos: {score['pos']}, Neu: {score['neu']}, Neg: {score['neg']}")
```

```python
###### Comparing sentiment scores for emoji-rich sentences before and after converting emojis into text descriptions using the `emoji` library.
import emoji
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Step 1: Create emoji-rich sentences
sentences = [
    "I'm so happy today! 😄🎉",
    "This is the worst day ever 😭💔",
    "Feeling relaxed and peaceful 🧘‍♀️✨",
    "Ugh, you're so annoying 😒🙄",
    "Just aced my test! 💯🔥"
]

# Step 2: Demojize sentences
demojized = [emoji.demojize(s) for s in sentences]

# Step 3: Initialize VADER
analyzer = SentimentIntensityAnalyzer()

# Step 4: Score both versions
for original, converted in zip(sentences, demojized):
    original_score = analyzer.polarity_scores(original)['compound']
    converted_score = analyzer.polarity_scores(converted)['compound']
    print(f"Original: {original} → {original_score}")
    print(f"Demojized: {converted} → {converted_score}")
    print("----")
```

```python
###### Testing how VADER scores capitalization and the addition of more emotion using exclamation marks 
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()

# Two versions
text1 = "I asbolutely love this!"
text2 = "I absolutely LOVE this!"
text3 = "I absolutely LOVE this!!!!!"

# Score both
score1 = analyzer.polarity_scores(text1)
score2 = analyzer.polarity_scores(text2)
score3 = analyzer.polarity_scores(text3)

print("Lowercase version:", score1)
print("All-caps version:", score2)
print("All-caps version with more exclamation marks:", score3)
```

```python
#### testing changes in VADER scores when emojis, exclamation marks and capitalization is added
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import pandas as pd

# Initialize VADER
analyzer = SentimentIntensityAnalyzer()

# Define the example sentences
examples = [
    # Positive
    "The movie was amazing",
    "The movie was amazing!!!",
    "The movie was AMAZING!!!! ❤️😍",
    
    # Negative
    "That movie was bad",
    "That movie was really bad!!",
    "THAT MOVIE WAS TERRIBLE 😡💀",
    
    # Neutral
    "It was okay",
    "It was okay... I guess",
    "It was OKAY 🤷‍♂️",
    
    # Mixed
    "Great visuals but weak plot",
    "GREAT visuals, but really weak plot!",
    "Great visuals 😍 but the plot was awful 😖",
    
    # Strong positive slang
    "So good!",
    "SOOO GOOD!!! 🔥🔥",
    "100% must-watch 💯👌"
]

# Analyze sentiment
results = []
for sentence in examples:
    scores = analyzer.polarity_scores(sentence)
    results.append({
        'Text': sentence,
        'Compound': scores['compound'],
        'Positive': scores['pos'],
        'Neutral': scores['neu'],
        'Negative': scores['neg']
    })

# Create DataFrame
vader_comparison_df = pd.DataFrame(results)

# Display
print(vader_comparison_df)

# Optional: Display as styled table (Jupyter only)
vader_comparison_df.style.background_gradient(cmap="coolwarm", subset=['Compound'])
```
